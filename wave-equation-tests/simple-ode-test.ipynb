# Solving an ODE

````python
from scipy.integrate import solve_ivp
import numpy as np
import matplotlib.pyplot as plt
````

* ### Newton's First Equation of Motion:  
    * $y(t) =\frac{-1}{2} gt^2 + v_0t + y_0$.
    * This is the analytics solution for the following differential equation. I will compare this to the result obtained using scipy's ODE solver.  

* ### Taking the derivative:
    * $\frac{dy}{dt} = -at + v_0$.
    * Initial Condition: $y(0) = 100$.
    * This is the equation passed into scipy's ```solve_ivp``` function.

````python
# derivative
def f(t, y):
    v0 = 5
    return -9.8*t + v0

# analytically determined solution
def yf(t):
    v0 = 5
    return -1/2 * 9.8 * t**2 + v0 * t + 100
    
# initial conditions
y0 = np.array([100])

# solve ode
sol = solve_ivp(f, (0,10), y0, t_eval=np.linspace(0,10,100))
y = sol.y
````

## Visualizing The Results

````python
plt.style.use("seaborn-pastel")
plt.title("Solutions")
plt.xlabel("time (seconds)")
plt.ylabel("vertical displacement (meters)")
plt.plot([i/10 for i in range(100)], [y[0][i] for i in range(100)], alpha=0.5, label="ODE Solution")
plt.plot([i/10 for i in range(100)], [yf(i) for i in np.linspace(0, 10, 100)], alpha=0.25, label="Analytic Solution")
plt.legend()
plt.show()
````

These functions overlap perfectly.
